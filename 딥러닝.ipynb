# CHAPTER 2. 퍼셉트론
###퍼셉트론
---
*   프랑크 로젠블라트(1957년)
*   신경망(딥러닝)의 기원이 되는 알고리즘

## 2.1 퍼셉트론이란?
---
*   다수의 신호를 입력으로 받아 하나의 신호를 출력
*   신호의 종류는 흐른다(1)/안 흐른다(0)
*   입력신호가 뉴런(노드)에 보내질 때 각각 고유한 가중치가 곱해짐
*   뉴런(노드)에서 보내온 신호의 총합이 정해진 한계를 넘어설 때만 1을 출력(이것을 뉴런을 활성화한다라고 표현하기도 함)
*   가중치가 클수록 해당 신호가 그만큼 더 중요

## 2.2 단순한 논리 회로
---
###2.2.1 AND 게이트
*   입력이 2개이다
*   둘 다 1일때만 1을 출력 나머지는 다 0을 출력
*   만약 입력값이 0.8/0.9라고 하더라도 1.0이 아니므로 0으로 처리

###2.2.2 NAND 게이트와 OR게이트
#### NAND=NOT AND(AND 게이트의 반대)
*   둘 다 1일때만 0을 출력 나머지는 다 1을 출력
*   AND 게이트의 입력값에 모두 부호를 반전하면 된다

#### OR
*   둘 다 0일때만 0을 출력 나머지는 다 1을 출력

#매개변수
---
*   매개변수는 인간이 설정해줌
*   기계학습의 문제는 이 매개변수 값을 정하는 작업을 컴퓨터가 자동으로 하게 함

## 2.3 퍼셉트론 구현하기
---
###2.3.1 간단한 구현부터




















```python
def AND(x1, x2):
  w1,w2, theta=0.5,0.5,0.7
  tmp=x1*w1+x2*w2
  if tmp<=theta:
    return 0
  elif tmp>theta:
    return 1
```

```python
print(AND(0,0)) #0을 출력
print(AND(1,0)) #0을 출력
print(AND(0,1)) #0을 출력
print(AND(1,1)) #1을 출력
```

### 2.3.2 가중치와 편향도입

```python
import numpy as np
x=np.array([0,1]) # 입력
w=np.array([0.5,0.5]) # 가중치
b=-0.7 # 편향
print(w*x)
print(np.sum(w*x)) #np.sum() 메서드는 입력한 배열에 담긴 모든 원소의 총합을 계산 
print(np.sum(w*x)+b) # 대략 -0.2 (부동소수점 수에 의한 연산 오차)
```

### 2.3.3 가중치와 편향 구현하기

```python
# 가중치와 편향을 도입한 AND 게이트
def AND(x1, x2):
  x=np.array([x1,x2]) # 입력
  w=np.array([0.5,0.5]) # 가중치
  b=-0.7 # 편향
  tmp=np.sum(w*x)+b
  if tmp<=0:
    return 0
  else:
    return 1
# 편향 : 한쪽으로 치우쳐 균형을 깬다(여기서 입력이 모두 0이어도 0이 아닌 편향값을 출력함)
# w1 & w2는 입력신호가 결과에 주는 영향력(중요도)을 조절하는 매개변수
# 편향은 뉴런이 얼마나 쉽게 활성화(결과로 1을 출력)하느냐를 조정하는 매개변수
# 예) 편향이 -0.1이라고 한다면 입력신호에 가중치를 곱한 값들의 합이 0.1을 초과할 때만 뉴런이 활성화됨
```

```python
def NAND(x1,x2):
  x=np.array([x1,x2])
  w=np.array([-0.5,-0.5]) # AND랑 가중치(w와 b)만 다르다
  b=0.7 # 원래 편향아니었나?
  tmp=np.sum(w*x)+b
  if tmp<=0:
    return 0
  else:
    return 1
def OR(x1,x2):
  x=np.array([x1,x2])
  w=np.array([0.5,0.5]) # AND랑 가중치(w와 b)만 다르다
  b=-0.2
  tmp=np.sum(w*x)+b
  if tmp<=0:
    return 0
  else:
    return 1
```

## 2.5 다층 퍼셉트론이 출동한다면
---
*   퍼셉트론으로는 XOR을 표현할 수 없음
*   퍼셉트론은 "층을 쌓아" 다층 퍼셉트론을 만들 수 있음

### 2.5.1 기존 게이트 조합하기
*   AND, OR, NAND를 조합해 XOR 게이트를 만들 수 있음
*   단층 퍼셉트론으로는 XOR 게이트를 표현할 수 없다(다층 퍼셉트론으로는 비선형 영역을 분리할 수 없다)
*   NAND 게이트와 OR 게이트를 지나서 출력된 값을 AND 게이트에 넣으면 XOR 게이트와 같은 값이 나옴

### 2.5.2 XOR 게이트 구현하기






```python
def XOR(x1,x2):
  s1=NAND(x1,x2)
  s2=OR(x1,x2)
  y=AND(s1,s2)
  return y
print(XOR(0,0)) #0을 출력
print(XOR(1,0)) #1을 출력
print(XOR(0,1)) #1을 출력
print(XOR(1,1)) #0을 출력
```

### 3.2.4 시그모이드 함수 구현하기

### 4.2.5 왜 손실 함수를 설정하는가?

### 4.3.1 미분
*   예시) 10분간 2km를 달렸다. 이때 속도는 2/10=0.2[km/분]이다. 1분에 0.2km만큼의 속도(변화)로 뛰었다. 
*   위의 예에서 "달린거리"가 "시간"에 대해 얼마나 변화했는가를 계산했다. ->평균속도를 구한 것
*   미분은 "특정 순간"의 변화량을 뜻한다.
*   시간을 가능한 한 줄여 1분에 달린 거리... 1초에 달린 거리... 0.1초에 달린 거리 이런 식으로 간격을 줄여 한 순간의 변화량(어느 순간의 속도)을 얻는 것

```python
# 나쁜 구현의 예
def numerical_diff(f, x):
  h=1e-50 # 0,00...1형태에서 소수점 아래 0이 50개 라는 의미->반올림 오차가 생김
  return (f(x-h)-f(x))/h
# 1. 미세한 값으로 10의 -4승을 사용
# 2. (x+h)와 (x-h)일 때의 함수 f의 차분을 계산하는 방법을 씀 -> 중심차분, 중앙차분
```

```python
def numerical_diff(f,x):
  h=1e-4
  return (f(x+h)-f(x-h))/(2*h)
```

#4.4 기울기
---
*   변수 2개의 편미분을 동시에 계산하고 싶을 때 -> 모든 변수의 편미분을 백터로 정리한 것을 기울기라 함

```python
def numerical_gradient(f,x): # f가 함수 x가 넘파이 배열 -> x의 각 원소에 대해서 수치 미분을 구함
  h=1e-4 #0.0001
  grad=np.zeros_like(x) # x와 형상(모양)이 같고 원소가 모두 0인 배열을 만듦
  for idx in range(x.size):
    tmp_val=x[idx]

    # f(x+h) 계산
    x[idx]=tmp_val+h
    fxh1=f(x)

    #f(x-h) 계산
    x[idx]=tmp_val-h
    fxh2=f(x)

    grad[idx]=(fxh1-fxh2)/(2*h)
    x[idx]=tmp_val # 값복원
  return grad

print(numerical_gradient(function_2, np.array([3.0,4.0]))) # 여기의 3.0 ,4.0같은 것들은 점을 뜻함
print(numerical_gradient(function_2, np.array([0.0,2.0])))
print(numerical_gradient(function_2, np.array([3.0,0.0])))
```

*   기울기 그림은 방향을 가진 화살표로 그려짐
*   한 점을 향하고 함수의 가장 낮은 장소(최솟값)를 가르킴 근데 또 반드시 그런 것은 아님 
*   "기울기가 가르키는 쪽은 각 장소에서 함수의 출력 값을 가장 크게 줄이는 방향" 아주 중요함
### 4.4.1 경사법(경사 하강법)
*   손실함수가 최솟값이 될 때의 매개변수 값을 찾기가 힘들다 -> 기울기 잘 이용해 최솟값(가능한 한 작은 값)을 찾으려는 것==경사법
*   주의할 점->각 지점에서 함수의 값을 낮추는 방안을 제시하는 지표가 기울기임
(기울기가 0인 것 1.극솟값 2.최솟값 3.안장점)
*   함수의 모양이 복잡하거나 찌그러졌다면 평평한 곳으로 파고들면서 고원(플래토)이라고 하는 학습이 진행되지 않는 정체기에 빠질 수 있음
*   경사법-> 현 위치에서 기울어진 방향으로 일정 거리만큼 이동하고 이동한 곳에서도 기울기를 구한다 그 후 다시 이동하면서 과정을 반복한다
*   7처럼 생긴 에타라는 기호는 갱신하는 양을 나타낸다 -> 이것을 학습률이라고 함
*   갱신단계를 반복하면서 함수값을 줄임 -> 변수가 늘어도 같은 식(편미분값)으로 갱신
*   학습률 값은 0.01이나 0.001같이 특정 값으로 미리 정해야 함 -> 일반적으로 너무 크거나 작으면 "좋은 장소"를 찾아가지 못함(학습률값을 변경하면서 진행)


```python
import numpy as np
def numerical_gradient(f,x): # f가 함수 x가 넘파이 배열 -> x의 각 원소에 대해서 수치 미분을 구함
  h=1e-4 #0.0001
  grad=np.zeros_like(x) # x와 형상(모양)이 같고 원소가 모두 0인 배열을 만듦
  for idx in range(x.size):
    tmp_val=x[idx]

    # f(x+h) 계산
    x[idx]=tmp_val+h
    fxh1=f(x)

    #f(x-h) 계산
    x[idx]=tmp_val-h
    fxh2=f(x)

    grad[idx]=(fxh1-fxh2)/(2*h)
    x[idx]=tmp_val # 값복원
  return grad
def gradient_descent(f,init_x, lr=0.01, step_num=100):
# f는 최적화하려는 함수 
# init_x는 초깃값
# lr은 학습률
# step_num은 경사법의 반복횟수
  x=init_x

  for i in range(step_num):
    grad=numerical_gradient(f,x) #함수의 기울기구함
    x-=lr*grad
  return x
def function_2(x):
  return x[0]**2+x[1]**2
init_x=np.array([-3.0, 4.0]) # 초기값
gradient_descent(function_2, init_x=init_x, lr=0.1,step_num=100)
```
